```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Position_Salaries.csv')
X=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

Comparing Linear Regression vs Polynomial Regression

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

```python
poly_lm_2 = LinearRegression()
poly_lm_2.fit(X_poly,y)
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lm.predict(X))
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Truth or Bluff(Linear Regression)')
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,poly_lm_2.predict(X_poly))
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.show()
```

```python
print(lm.predict([[6.5]]))
```

```python
print(poly_lm_2.predict([[poly_reg.fit_transform(6.5)]]))
```
